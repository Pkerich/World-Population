## Import Libraries and Data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Administrator\Desktop\Male and Female.csv")
```

```python
df.rename(columns= {"Gender" : "Year"}, inplace=True)
df.head()
```

```python
df.sample(n=20).sort_index(axis=0).head()
```

Communicate Results in a plot

```python
df.sample(n=20).sort_index(axis=0).plot(
    kind="bar", 
    x="Year", 
    y=["male", "female", "Grand Total"], 
    title="Distribution of Population by Gender",
    ylabel="Population in Billions",
    figsize=(20, 10))
```

```python
df.plot.line(x="Year", ylabel="Population in Billions", title="Global Population Distribution by Gender", figsize=(10,8))
```

```python
df.plot.line(x="Year", ylabel="Population in Billions", title="Global Population Distribution by Gender", figsize=(10,8), subplots=True)
```
